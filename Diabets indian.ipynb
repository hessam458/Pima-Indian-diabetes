```python
import numpy as np
import pandas as pd
data = pd.read_csv('diabetes.csv')
```

```python

data.describe()
```

```python
mean_diabetes = data[(data['Insulin'] != 0) & (data['Outcome']==1)]['Insulin'].mean()
data.loc[(data['Insulin']==0) & (data['Outcome']==1),'Insulin'] = mean_diabetes
mean_nondiabetes =  data[(data['Insulin'] != 0) & (data['Outcome']==0)]['Insulin'].mean()
data.loc[(data['Insulin']==0) & (data['Outcome']==0),'Insulin'] = mean_nondiabetes

median_bloodpressure = data[data['BloodPressure']!=0]['BloodPressure'].median()
data.loc[data['BloodPressure']==0,'BloodPressure'] = median_bloodpressure

mean_BMI = data[data['BMI']!=0]['BMI'].mean()
data.loc[data['BMI']==0,'BMI'] = mean_BMI

median_skinthickness = data[data['SkinThickness']!=0]['SkinThickness'].median() 
data.loc[data['SkinThickness']==0,['SkinThickness']] = median_skinthickness
```
